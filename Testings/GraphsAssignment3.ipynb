```python
def generateGraphForGameReport(GameReport,gameNumber,supTitle):
  Turns = range(GameReport['Player1']['turns'])
  BoardsP1 = GameReport['Player1']['boardsPerTurn']
  BoardsP2 = GameReport['Player2']['boardsPerTurn']
  PruningsP1 = GameReport['Player1']['pruningsPerTurn']
  PruningsP2 = GameReport['Player2']['pruningsPerTurn']
  Times=[GameReport['Player1']['timeTook'],GameReport['Player2']['timeTook']]
  if len(PruningsP1)>len(PruningsP2):
    PruningsP2.append(0)
  elif len(PruningsP1)<len(PruningsP2):
    PruningsP1.append(0)  
  if len(BoardsP1)>len(BoardsP2):
    BoardsP2.append(0)
  elif len(BoardsP1)<len(BoardsP2):
    BoardsP1.append(0)  
  labels=["Player1","Player2"]
  bar_colors = ['tab:red', 'tab:blue']
  fig, ax = plt.subplots(1, 3,figsize=(12, 4))
  ax[0].plot(Turns, BoardsP1, label="Player1")
  ax[0].plot(Turns, BoardsP2, label="Player2")
  ax[0].set_title('Boards Found Each Turn on Game '+str(gameNumber))
  ax[0].legend()

  ax[1].plot(Turns, PruningsP1, label="Player1")
  ax[1].plot(Turns, PruningsP2, label="Player2")
  ax[1].set_title('Prunings Done Each Turn on Game '+str(gameNumber))
  ax[1].legend()

  ax[2].bar(["Player 1","Player2"], Times, color=bar_colors)
  ax[2].set_title('Time Used on Game '+str(gameNumber)+' [s]')
  ax[2].legend()
  fig.suptitle(supTitle,y=1.05,fontweight="bold",fontsize=15)
#ax.plot(values, cat, label="cat")
generateGraphForGameReport(GameReport1,1,"Heuristic 1 Depth 3 vs Heuristic 2 Depth 3\n Won Player 1")
```
